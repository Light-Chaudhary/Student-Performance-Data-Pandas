```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
print(df.isnull().sum())
missing_values = df[df.isnull().any(axis = 1)]
print(missing_values) 
df = df.dropna(subset = ["Age", "Embarked"])
print(df)
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df["Name"] = df["Name"].str.lower()
df["Last_Name"] = df["Name"].str.split(",").str[0].str.strip()
find_mrs_miss = df["Name"].str.contains("miss|mrs" , na = False, case = False)
print(find_mrs_miss)
df = df.drop_duplicates()
df["Survived"] = df["Survived"].astype(str)
df["Fare"] = df["Fare"].astype(int)
df["Is_child"] = df["Age"].apply(lambda x : 1 if x < 18 else 0)
sex_map = {"male": 0 , "female": 1 }
df["Sex"] = df["Sex"].map(sex_map)
replacing  = {"S": "Southampton", "C": "Cherbourg", "Q": "Queenstown"}
df["Embarked"] = df["Embarked"].replace(replacing)
df["Cabin"] = df["Cabin"].fillna("Unknown")
df
```
